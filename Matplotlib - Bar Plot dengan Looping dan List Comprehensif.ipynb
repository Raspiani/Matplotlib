### Matplotlib - Bar Plot Part 2

```python
#masukan modul yang digunakan 
import matplotlib.pyplot as plt, numpy as np, random as rd, pandas as pd
%matplotlib inline
```

```python
#masukan data dalam hal ini misalnya data terdiri dari data, register, dan pets
data = [10,25,100,40]
register = 0,1,2,3
pets = "CAT", "DOG", "HEDGEHOG", "BUNNY"
```

```python
#memvisualisasikan data dengan matplotlib

#membuat ukuran frame
plt.figure(figsize = (8,4))

#memberikan judul garfik
plt.title("FAVOURITE PETS", fontsize = 20)

#membuat grafik bar chart dengan matplotlib
b = plt.bar(register, data, width = 0.2, color = ["r", "g", "b", "m"])
plt.xticks(register, pets)

#memberikan keterangan lebih atau legend 
plt.legend(b, pets, fontsize = 12)

#menampilkan hasil grafik
plt.show()
```

```python
#memvisualisasikan grafik dengan bar chart menggunakan width = 1
plt.figure(figsize = (8,4))

plt.title("FAVOURITE PETS", fontsize = 20)

b = plt.bar(register, data, width = 1, color = ["r", "g", "b", "m"])
plt.xticks(register, pets)

plt.legend(b, pets, fontsize = 12)

plt.show()
```

```python
#memvisualisasikan grafik dengan bar chart menggunakan width = 0.8
plt.figure(figsize = (8,4))

plt.title("FAVOURITE PETS", fontsize = 20)

b = plt.bar(register, data, width = 0.8, color = ["r", "g", "b", "m"])
plt.xticks(register, pets)

plt.legend(b, pets, fontsize = 12)

plt.show()
```

```python
#memvisualisasikan bar chart secara horizontal dengan barh 
#karena barh maka height dan yticks
#dalam hal ini nilai height = 0.8
plt.figure(figsize = (8,4))

plt.title("FAVOURITE PETS", fontsize = 20)

b = plt.barh(register, data, height = 0.8, color = ["r", "g", "b", "m"])
plt.yticks(register, pets)

plt.legend(b, pets, fontsize = 12)

plt.show()
```

```python
#mengenerate data
rd.seed(10)

drinks = [
    rd.sample((range(1,20)), 5),
    rd.sample((range(1,20)), 5),
    rd.sample((range(1,20)), 5) ]

drinks
```

```python
#memvisualisasikan dengan bar chart
x = np.arange(5)

days = "Mon", "Tue", "Wed", "Thur", "Fri"

plt.bar(x + 0.00, drinks[0], color = "b", width = 0.25, label = "Latte")
plt.bar(x + 0.25, drinks[1], color = "r", width = 0.25, label = "Capuccino")
plt.bar(x + 0.5, drinks[2], color = "g", width = 0.25, label = "Tea")

plt.title("HOT DRINKS", fontsize = 20)
plt.xticks(x, days)
plt.legend(fontsize = 15)

plt.show()
```

```python
#memvisualisasikan dengan bar chart
#plt.legend(fontsize = 15, loc = (1.05, 0.4))
#legend(fontsize = 15, loc = (1.05, 0.4))
x = np.arange(5)

days = "Mon", "Tue", "Wed", "Thur", "Fri"

plt.bar(x + 0.00, drinks[0], color = "b", width = 0.25, label = "Latte")
plt.bar(x + 0.25, drinks[1], color = "r", width = 0.25, label = "Capuccino")
plt.bar(x + 0.5, drinks[2], color = "g", width = 0.25, label = "Tea")

plt.title("HOT DRINKS", fontsize = 20)
plt.xticks(x, days)

#loc = (1.05, 0.4)
plt.legend(fontsize = 15, loc = (1.05, 0.4))

plt.show()
```

```python
#memvisualisasikan dengan barh
#karena barh maka height dan yticks
x = np.arange(5)

days = "Mon", "Tue", "Wed", "Thur", "Fri"

plt.barh(x + 0.00, drinks[0], color = "b", height = 0.25, label = "Latte")
plt.barh(x + 0.25, drinks[1], color = "r", height = 0.25, label = "Capuccino")
plt.barh(x + 0.5, drinks[2], color = "g", height = 0.25, label = "Tea")

plt.title("HOT DRINKS", fontsize = 20)
#yticks
plt.yticks(x, days)
plt.legend(fontsize = 15, loc = (1.05, 0.4))

plt.show()
```

```python
#contoh kesalahan jika xticks ditulis yticks
#memvisualisasikan dengan bottom
x = np.arange(5)

days = "Mon", "Tue", "Wed", "Thur", "Fri"

plt.bar(x, drinks[0], bottom = np.sum(drinks[:0], axis=0), label = "Latte")
plt.bar(x, drinks[1], bottom = np.sum(drinks[:1], axis=0), label = "Capuccino")
plt.bar(x, drinks[2], bottom = np.sum(drinks[:2], axis=0), label = "Tea")

plt.title("HOT DRINKS", fontsize = 20)
plt.yticks(x, days)
plt.legend(fontsize = 15, loc = (1.05, 0.4))

plt.show()
```

```python
#xticks
#memvisualisasikan dengan bottom
x = np.arange(5)

days = "Mon", "Tue", "Wed", "Thur", "Fri"

plt.bar(x, drinks[0], bottom = np.sum(drinks[:0], axis=0), label = "Latte")
plt.bar(x, drinks[1], bottom = np.sum(drinks[:1], axis=0), label = "Capuccino")
plt.bar(x, drinks[2], bottom = np.sum(drinks[:2], axis=0), label = "Tea")

plt.title("HOT DRINKS", fontsize = 20)
plt.xticks(x, days)
plt.legend(fontsize = 15, loc = (1.05, 0.4))

plt.show()
```
