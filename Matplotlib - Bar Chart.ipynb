### Matplotlib Bar Chart

```python
#memasukan modul yang digunakan
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#memasukan data, misalnya dalam hal ini company, reveneu dan profit
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
profit = [48, 2, 34, 12]
```

```python
#memberikan nomor urutan untuk company
urutan_company_x = np.arange(len(company))
urutan_company_x
```

```python
#menggambarkan bar plot di matplotlib
plt.bar(urutan_company_x, revenue)
```

```python
#memberikan label untuk sumbu x dan y 
plt.xticks(urutan_company_x, company)
plt.ylabel("revenue(bln)")

plt.bar(urutan_company_x, revenue)
```

```python
#menggambarkan bar plot di matplotlib dengan title
plt.title("US Tech Stocks")

plt.xticks(urutan_company_x, company)
plt.ylabel("revenue(bln)")

plt.bar(urutan_company_x, revenue)
```

```python
#menambahkan legend()
plt.title("US Tech Stocks")

plt.xticks(urutan_company_x, company)
plt.ylabel("revenue(bln)")

plt.bar(urutan_company_x, revenue, label = "Revenue")

plt.legend()
```

```python
#menambahkan bar(urutan_company_x, label = "Profit")
plt.title("US Tech Stocks")

plt.xticks(urutan_company_x, company)
plt.ylabel("revenue(bln)")

plt.bar(urutan_company_x, revenue, label = "Revenue")
plt.bar(urutan_company_x, profit, label = "Profit")

plt.legend()
```

```python
#menambahkan urutan_company_x dengan -+ 0.2
plt.title("US Tech Stocks")

plt.xticks(urutan_company_x, company)
plt.ylabel("revenue(bln)")

plt.bar(urutan_company_x - 0.2, revenue, label = "Revenue")
plt.bar(urutan_company_x + 0.2, profit, label = "Profit")

plt.legend()
```

```python
#menambahkan width = 0.4
plt.title("US Tech Stocks")

plt.xticks(urutan_company_x, company)
plt.ylabel("revenue(bln)")

plt.bar(urutan_company_x - 0.2, revenue, width = 0.4, label = "Revenue")
plt.bar(urutan_company_x + 0.2, profit, width = 0.4, label = "Profit")

plt.legend()
```

```python
#mengubah letak nama company menjadi di sumbu y
urutan_company_y = np.arange(len(company))
urutan_company_y
```

```python
#mengubah xticks menjadi yticks
plt.title("US Tech Stocks")

plt.yticks(urutan_company_y, company)

plt.barh(urutan_company_y - 0.2, revenue, label = "Revenue")
plt.barh(urutan_company_y + 0.2, profit, label = "Profit")

plt.legend()
```
