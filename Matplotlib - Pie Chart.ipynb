<h3 id="Matplotlib---Pie-Chart">Matplotlib - Pie Chart</h3>

<pre>
<code>Berikut ini adalah langkah-langkah untuk membuat sebuah pie chart dengan modul matplotlib di python. Dimana data 
yang diberikan yaitu:

exp_vals = [1400,600,300,410,250]
exp_labels = [&quot;Home Rent&quot;, &quot;Food&quot;, &quot;Phone/Internet Bill&quot;, &quot;Car&quot;, &quot;Other Utilities&quot;]

untuk fitur lebih lengkap terkait matplotlib dapat mengunjungi <a href="https://matplotlib.org/gallery/index.html">https://matplotlib.org/gallery/index.html
</a></code></pre>


```python
#memasukan modul yang digunakan 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#masukan data dalam hal ini exp_vals = x, dan exp_labels = y
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
```

```python
#membuat pie chart dengan matplotlib
plt.pie(exp_vals, labels = exp_labels)
```

```python
#pie chart dengan axis("equal") dan radius = 1.5
plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels, radius = 1.5)
plt.show()
```

```python
#pie chart dengan axis("equal") dan radius = 1.5
#autopct = "%0.0f%%"
plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels, radius = 1.5, autopct = "%0.0f%%")
plt.show()
```

```python
#pie chart dengan axis("equal") dan radius = 1.5
#autopct = "%0.1f%%"
plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels, radius = 1.5, autopct = "%0.1f%%")
plt.show()
```

```python
#pie chart dengan axis("equal") dan radius = 1.5
#autopct = "%0.1f%%"
#shadow = True
#explode=[0,0.5,0,0,0]
#untuk men-slice seperti pizza 
plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels, radius = 1.5, autopct = "%0.1f%%", shadow=True, explode=[0,0.5,0,0,0])
plt.show()
```

```python
#pie chart dengan axis("equal") dan radius = 1.5
#autopct = "%0.1f%%"
#shadow = True
#explode=[0,0.1,0.1,0,0]
#untuk men-slice seperti pizza 
plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels, radius = 1.5, autopct = "%0.1f%%", shadow=True, explode=[0,0.1,0.1,0,0])
plt.show()
```

```python
#pie chart dengan axis("equal") dan radius = 1.5
#autopct = "%0.1f%%"
#shadow = True
#explode=[0,0.1,0.1,0,0]
#startangle = 45
#untuk men-slice seperti pizza 
plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels, radius = 1.5, autopct = "%0.1f%%", shadow=True, explode=[0,0.1,0.1,0,0], startangle = 45)
plt.show()
```

```python
#pie chart dengan axis("equal") dan radius = 1.5
#autopct = "%0.1f%%"
#shadow = True
#explode=[0,0.1,0.2,0,0]
#startangle = 180
#untuk men-slice seperti pizza 
plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels, radius = 1.5, autopct = "%0.1f%%", shadow=True, explode=[0,0.1,0.2,0,0], startangle = 180)
plt.show()
```
