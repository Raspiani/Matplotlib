## Visualisasi Data dengan Matplotlib

### Pengenalan Matplotlib

    Matplotlib adalah salah satu modul populer di python yang digunakan untuk visualisasi data. Matplotlib bisa dibaca dan dipelajari lebih lengkap di https://matplotlib.org/

```python
#masukan modul yang digunakan 
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#membuat nilai x dengan modul numpy dengan rentang (-6,6,200)
x = np.linspace(-6,6,200)

cos = np.cos(x)

plt.plot(x,cos)
```

```python
x = np.linspace(-6,6,200)

sin = np.sin(x)

plt.plot(x,sin)
```

```python
x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos)
plt.plot(x, sin)
```
