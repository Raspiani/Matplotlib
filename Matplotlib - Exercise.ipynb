### Matplotlib - Exercise

    1. Gambarkan 4 subplot dari sinx, cosx, -sinx, -cosx
    2. Tambahkan judul untuk masing-masing subplot tersebut, gunakan linestyle yang berbeda untuk masing-masing subplot
    3. Tambahkan label teks untuk satu atau lebih subplot tersebut.

### Jawaban Exercise

```python
#memasukan modul yang digunakan
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(-6,6,180)
cos = np.cos(x)
sin = np.sin(x)
```

```python
plt.figure(figsize = (12,6))

plt.subplot(221)
plt.plot(x, cos, color = "red", linewidth = 2.6, linestyle = ':')
plt.title('cos x')
plt.grid(True)
plt.text(-5,-0.3, "Ini akan melintasi subplot", rotation = 10, fontsize = 15)

plt.subplot(222)
plt.plot(x, sin, color = "green", linewidth = 3.0, linestyle = '-.')
plt.title('sin x')
plt.grid(True)
plt.text(-5,-0.3, "Ini akan melintasi subplot", rotation = 10, fontsize = 15)

plt.subplot(223)
plt.plot(x, -cos, color = 'blue', linewidth = 3.5, linestyle = '--')
plt.title('-cos x')
plt.grid(True)

plt.subplot(224)
plt.plot(x, -sin, color = 'purple', linewidth = 4.0, linestyle = '-')
plt.title('-sin x')
plt.grid(True)
```
