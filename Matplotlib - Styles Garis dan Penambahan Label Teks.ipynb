## Matplotlib - Styles Garis dan Penambahan Label Teks

```python
#memasukan modul yang digunakan 
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(-10,10,150)

cos = np.cos(x)
sin = np.sin(x)
```

```python
#style garis yaitu :, -. , --, -
```

```python
plt.plot(x, cos, color = 'red', linewidth = 3.8, linestyle = ':')
```

```python
plt.plot(x, sin, linewidth = 3.5, linestyle = '-.')
```

```python
plt.plot(x, sin, linewidth = 3.5, color = 'green', linestyle = '--')
plt.plot(x, cos, linewidth = 3.5, color = 'red', linestyle = '-')
```

```python
plt.plot(x, cos, color = 'green', linewidth = 3.2)

plt.text(-4,-0.3, 'Teks ini akan ditampilkan pada plot', rotation = 20, fontsize = 14)
```

```python
plt.plot(x, cos, color = 'green', linewidth = 3.2)

plt.text(-7, 0, 'Teks ini akan ditampilkan pada plot', rotation = 35, fontsize = 14)
```

```python
plt.plot(x, cos, color = 'green', linewidth = 3.2)

plt.text(-8,0, 'Teks ini akan ditampilkan pada plot', fontsize = 14)
```

```python
plt.plot(x, cos, color = 'green', linewidth = 3.2)

plt.text(-4, -0.3, 'Teks ini akan ditampilkan pada plot', rotation = -20, fontsize = 14)
```

```python
plt.subplot(211)
plt.plot(x, cos, color = 'red', linewidth = 2.8)
plt.text(-5, -0.5, 'Ini akan menjadi subplot', rotation = 11, fontsize = 14)

plt.subplot(212)
plt.plot(x, sin, color = 'green', linewidth = 3.8)
plt.text(-5, -0.5, 'Ini akan menjadi subplot', rotation = -11, fontsize = 14)
```

```python
plt.subplot(121)
plt.plot(x, cos, color = 'red', linewidth = 2.8)
plt.text(-5, -0.5, 'Ini akan menjadi subplot', rotation = 11, fontsize = 14)

plt.subplot(122)
plt.plot(x, sin, color = 'green', linewidth = 3.8)
plt.text(-5, -0.5, 'Ini akan menjadi subplot', rotation = -11, fontsize = 14)
```

```python
plt.figure(figsize = (10,6))

plt.subplot(121)
plt.plot(x, cos, color = 'red', linewidth = 2.8)
plt.text(-5, -0.5, 'Ini akan menjadi subplot', rotation = 11, fontsize = 14)

plt.subplot(122)
plt.plot(x, sin, color = 'green', linewidth = 3.8)
plt.text(-5, -0.5, 'Ini akan menjadi subplot', rotation = -11, fontsize = 14)
```
