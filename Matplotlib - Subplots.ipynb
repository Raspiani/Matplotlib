### Matplotlib - Subplots

    Misalnya subplot, jumlah dari baris, jumlah kolom, jumlah gambar dsb. Contoh subplot(211) = subplot(2,1,1)

```python
#masukan modul yang digunakan 
import numpy as np 

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.subplot(211)

x = np.linspace(-10,10,150)
cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = 'red', linewidth = 2.4)

plt.subplot(212)

plt.plot(x, sin, color = 'green', linewidth = 3.4)
```

```python
plt.figure(figsize = (10,5))
plt.subplot(211)

x = np.linspace(-10,10,150)
cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = 'red', linewidth = 2.4)

plt.subplot(212)

plt.plot(x, sin, color = 'green', linewidth = 3.4)
```

```python
plt.axes([1,1.2,1,1]) #left, bottom, width, hight

x = np.linspace(-10,10,150)
cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = 'red', linewidth = 2.4)

plt.axes([1,1,1,1])

plt.plot(x, sin, color = 'green', linewidth = 3.4)
```
