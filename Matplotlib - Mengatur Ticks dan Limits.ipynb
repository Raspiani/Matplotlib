### Matplotlib - Mengatur Ticks dan Limits

```python
#memasukan modul yang digunakan 
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.figure(figsize = (10,5))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = "red", linewidth = 2.5)
plt.plot(x, sin, color = "green", linewidth = 3.5)
```

```python
plt.figure(figsize = (10,5))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = "red", linewidth = 2.5)
plt.plot(x, sin, color = "green", linewidth = 3.5)

plt.xlim(-9,9)
plt.ylim(-1.4,1.4)
```

```python
plt.figure(figsize = (10,5))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = "red", linewidth = 2.5)
plt.plot(x, sin, color = "green", linewidth = 3.5)

plt.xlim(x.min()+0.5, x.max()+0.5)
```

```python
plt.figure(figsize = (10,5))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = "red", linewidth = 2.5)
plt.plot(x, sin, color = "green", linewidth = 3.5)

plt.xlim(x.min()-2, x.max()+2)
```

```python
plt.figure(figsize = (10,5))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = "red", linewidth = 2.5)
plt.plot(x, sin, color = "green", linewidth = 3.5)

plt.xlim(x.min()-2, x.max()+2)
plt.ylim(sin.min()*1.3, sin.max()*1.3)
```

```python
plt.figure(figsize = (10,5))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = "red", linewidth = 2.5)
plt.plot(x, sin, color = "green", linewidth = 3.5)

plt.xticks([-6,-3,0,1,3,6])
plt.yticks([-1,-0.4,0,0.5,0.8,1])
```
