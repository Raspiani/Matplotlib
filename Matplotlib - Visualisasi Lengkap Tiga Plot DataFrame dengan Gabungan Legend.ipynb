## Matplotlib - Visualisasi Lengkap Tiga Plot DataFrame dengan Gabungan Legend

    Berikut ini visualisasi lengkap tiga plot dari dataFrame menggunakan modul matplotlib 

```python
#memasukan modul yang digunakan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
#nama data
file = "C:/Users/RASPIANI/3. Data/Book2.xlsx"

#memasukan dan membaca data
data1 = pd.read_excel(file, sheet_name = "_7")
data1.head()
```

```python
#membuat  bingkai / frame dengan ukuran 20 x 10
plt.figure(figsize=(20,10))

#membuat format di sumbu y menjadi angka x 10 pangkat 7
plt.rcParams['text.usetex'] = True

#memdefinisikan nilai sumbu x dan y
x = data1['Peserta']
y1 = data1['NA AAN']
y2 = data1['NA PUC']
y3 = data1['NA EAN']

#plot nilai akhir AAN
plt.plot(x, y1, color = 'black', label='NA AAN')
plt.plot(x, y1, '^', color = 'black', alpha=1, markersize=11)

#plot nilai akhir PUC dengan skala y ditambah 2000000
plt.plot(x, y2, color = 'black', label='NA PUC')
plt.plot(x, y2, 's', color = 'black', alpha=1, markersize=11)

#plot nilai akhir EAN 
plt.plot(x, y3-2000000, color = 'black', label='NA EAN')
plt.plot(x, y3-2000000, 'o', color = 'black', alpha=1, markersize=11)

#membuat legend atau keterangan 
legend_elements = [Line2D([0], [0], marker = '^', color='black', label='NA AAN'), 
                   Line2D([0], [0], marker = 's', color='black', label='NA PUC'),
                   Line2D([0], [0], marker = 'o', color='black', label='NA EAN')]

plt.legend(title = '', fontsize = 12, handles = legend_elements)

#membuat urutan angka di sumbu x
plt.xticks(np.arange(54), fontsize = 15)

#menambah ukuran angka di sumbu y
plt.yticks(fontsize = 15)

#membuat label di sumbu x
plt.xlabel("Peserta ke-", fontsize = 20)

#membuat label di sumbu y
plt.ylabel("Nilai Akumulasi Iuran Normal", fontsize = 20)

#membuat judul grafik
plt.title("Grafik Nilai Akumulasi Iuran Normal dengan i = 7 \%", fontsize = 25)

#menyimpan hasil grafik menjadi png dan pdf
for gambar in ["png", "pdf"]:
    print("Grafik Nilai Akumulasi Iuran Normal dengan i = 7 persen.%s" % (gambar,))
    plt.savefig("Grafik Nilai Akumulasi Iuran Normal dengan i = 7 persen.%s" % (gambar,), bbox_inches="tight")

plt.show()
```
