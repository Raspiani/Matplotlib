### Matplotlib - Penambahan Text

```python
#memasukan modul yang digunakan
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(-10,10,150)
cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = 'red', linewidth = 3.0)
```

```python
plt.plot(x, cos, color = "red", linewidth = 3.0)

plt.xlabel("X-Axis")
plt.ylabel("cos x")
```

```python
plt.plot(x, cos, color = "red", linewidth = 3.0)

plt.xlabel("X-Axis")
plt.ylabel("cos x")
plt.title("Grafik dari x vs cos x")
```

```python
plt.plot(x, cos, color = "red", linewidth = 3.0)

plt.xlabel("x-Axis", fontsize = 15, color = "green")
plt.ylabel("cos x", fontsize = 15, color = "blue")
plt.title("Grafik dari x vs cos x")
```

```python
plt.plot(x, cos, color = "red", linewidth = 3.0)

plt.xlabel("x-Axis", fontsize = 15, color = "green")
plt.ylabel("cos x", fontsize = 15, color = "blue")
plt.title("Grafik dari x vs cos x")
plt.grid(True)
```

```python
plt.plot(x, cos, color = "red", linewidth = 3.0)

plt.xlabel("x-Axis", fontsize = 15, color = "green")
plt.ylabel("cos x", fontsize = 15, color = "blue")
plt.title("Grafik dari x vs cos x")
plt.grid(True)

plt.annotate("Ini adalah salah satu nilai maksimum", xy = (0,1), \
             xytext = (5, 0.5), arrowprops = dict(facecolor = "black", shrink = 0.09))
```

```python
plt.plot(x, cos, color = "red", linewidth = 3.0)

plt.xlabel("x-Axis", fontsize = 15, color = "green")
plt.ylabel("cos x", fontsize = 15, color = "blue")
plt.title("Grafik dari x vs cos x")
plt.grid(True)

plt.annotate("Ini adalah salah satu nilai maksimum", xy = (0,1), \
             xytext = (2.5, 1.5), arrowprops = dict(facecolor = "black", shrink = 0.1))
```
