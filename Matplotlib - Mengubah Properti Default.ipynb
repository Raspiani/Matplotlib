### Matplotlib - Mengubah Properti Default

```python
#memasukan modul yang digunakan
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos)
plt.plot(x, sin)
```

```python
plt.figure(figsize = (10,5))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos)
plt.plot(x, sin)
```

```python
plt.figure(figsize = (10,6))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos)
plt.plot(x, sin)
```

```python
plt.figure(figsize = (10,6))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = 'red', linewidth = 2.0)
plt.plot(x, sin)
```

```python
plt.figure(figsize = (10,6))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = 'red', linewidth = 2.0)
plt.plot(x, sin, color = 'green', linewidth = 3.0)
```

```python
plt.figure(figsize = (10,6))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = 'red', linewidth = 2.0)
plt.plot(x, sin, color = 'green', linewidth = 3.0)

plt.xticks(np.linspace(-6,6,12))
```

```python
plt.figure(figsize = (10,6))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = 'red', linewidth = 2.0)
plt.plot(x, sin, color = 'green', linewidth = 3.0)

plt.xticks(np.linspace(-6,6,12))
plt.yticks(np.linspace(-1,1,6))
```

```python
plt.figure(figsize = (10,6))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = 'red', linewidth = 2.0)
plt.plot(x, sin, color = 'green', linewidth = 3.0)

plt.xticks(np.linspace(-6,6,12))
plt.yticks(np.linspace(-1,1,6))

plt.xlim(-3,3)
```

```python
plt.figure(figsize = (10,6))

x = np.linspace(-6,6,200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = 'red', linewidth = 2.0)
plt.plot(x, sin, color = 'green', linewidth = 3.0)

plt.xticks(np.linspace(-6,6,12))
plt.yticks(np.linspace(-1,1,6))

plt.xlim(-8,8)
plt.ylim(-1.8, 1.8)
```
